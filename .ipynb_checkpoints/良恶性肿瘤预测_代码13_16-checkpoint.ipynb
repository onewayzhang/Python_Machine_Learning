```python
import pandas as pd
import numpy as np

column_names=['Sample Code number','Clump Thickness','Uniformity of Cell Size','Uniformity of Cell \
Shape','Marginal Adhersion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal\
 Nucleoli','Mitoses','Class']

data=pd.read_csv('E:/my_python_lessons/breast-cancer-wisconsin.data',names=column_names)

#将？替换成标准缺失值表示
data=data.replace(to_replace='?',value=np.nan)
data=data.dropna(how='any')
data.shape
```

```python
data
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data[column_names[1:10]], data[column_names[10]], test_size=0.25,\
                                            random_state=33)
```

代码15：使用线性分类模型（逻辑斯蒂回归和随机参数下降）从事良/恶性肿瘤预测


```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
    
ss=StandardScaler() #标准化数据，保证每个维度的特征数据均值为0，方差为1
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
ss
```

```python
X_train
```

```python
#初始化LogisticRegression 和 SGDClassifier
lr=LogisticRegression()
sgdc=SGDClassifier()
lr.fit(X_train,y_train)
```

```python
#使用训练好的模型lr对X_test进行预测
lr_y_predict=lr.predict(X_test)
```

```python
#使用SGDClassifier中的fit函数
sgdc.fit(X_train,y_train)
sgdc_y_predict=sgdc.predict(X_test)
```

```python
sgdc_y_predict
```

接下来进行性能测评

使用线性分类模型从事良/恶性肿瘤预测任务的性能分析

```python
from sklearn.metrics import classification_report

print ('Accuracy of LR Classifier:',lr.score(X_test,y_test))
```

```python
print (classification_report(y_test,lr_y_predict,target_names=['Benign','Malignant']))
```

```python
print ('Accuracy of SGD Classifier:',sgdc.score(X_test,y_test))
print (classification_report(y_test,sgdc_y_predict,target_names=['Benign','Malignant']))
```
