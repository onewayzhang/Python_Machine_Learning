# 构建测试集中正负分类样本

## 提取负样本的特征值

```python
''' 从df_test 中提取出 df_test['Type']==0的行，并进一步提取出想要的列 '''

df_test_negative=df_test[df_test['Type']==0][ ['Clump Thickness','Cell Size' ] ]
df_test_negative
```

## 提取正样本的特征值

```python
df_test_positive=df_test[df_test['Type']==1][ ['Clump Thickness','Cell Size' ] ]
df_test_positive.info()
```

# 导入matplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

## 绘制良性肿瘤样本点，为红色的圈

```python
plt.scatter(  df_test_negative['Clump Thickness'], df_test_negative['Cell Size'], marker='o', s=200, 
            c='red' )
```

## 绘制恶性肿瘤样本点，为黑色的x

```python
plt.scatter(  df_test_positive['Clump Thickness'], df_test_positive['Cell Size'], marker='x', s=150, 
            c='black' )
```

## 绘制x,y轴的说明

```python
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
```

```python
plt.show()
```

```python
import numpy as np
```

```python
intercept=np.random.random([1])
print(intercept)  # 截距
```

```python
coef=np.random.random([2])
print(coef)  #系数
```

## 绘制一条随机直线

```python
lx=np.arange(0,12)
ly=(-intercept-lx*coef[0])/coef[1]

plt.plot(lx,ly,c='yellow')
```

## 绘制图1-3

```python
plt.scatter(  df_test_negative['Clump Thickness'], df_test_negative['Cell Size'], marker='o', s=200, 
            c='red' )
plt.scatter(  df_test_positive['Clump Thickness'], df_test_positive['Cell Size'], marker='x', s=150, 
            c='black' )
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
```

```python
'''  注意： 所有的plt.scatter 和 .plot 等运行完毕后， 再plt.show(), 则所有绘制内容出现在一张图上'''
plt.show()
```

# 导入逻辑斯蒂回归分类器

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

## 使用训练集的前10条样本学习直线的系数和截距

```python
lr.fit( df_train.ix[ :10, [1,2] ], df_train.ix[ :10, 'Type' ] )
```

```python
print( 'Testing accuracy(10 training samples): ', lr.score(df_test.ix[:, [1,2]] ,df_test.ix[:, 3]   )  )
```

```python
intercept10=lr.intercept_
coef10=lr.coef_[0]     #  由二维数组得到一维数组
coef10
```

```python
ly=( -intercept10-lx*coef10[0] )/coef10[1]
```

## 绘制图1-4

```python
plt.plot(lx,ly,c='green')
plt.scatter(  df_test_negative['Clump Thickness'], df_test_negative['Cell Size'], marker='o', s=200, 
            c='red' )
plt.scatter(  df_test_positive['Clump Thickness'], df_test_positive['Cell Size'], marker='x', s=150, 
            c='black' )
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')

plt.show()
```

## 使用所有训练样本学习直线的系数和截距

```python
lr=LogisticRegression()
```

```python
lr.fit( df_train.ix[ :, [1,2] ], df_train.ix[ :, 'Type' ] )
print( 'Testing accuracy(10 training samples): ', lr.score(df_test.ix[:, [1,2]] ,df_test.ix[:, 3]   )  )
```

```python
intercept_all=lr.intercept_
coef_all=lr.coef_[0]     #  由二维数组得到一维数组
coef_all
```

```python
ly=( -intercept_all-lx*coef_all[0] )/coef_all[1]
```

## 绘制图1-5

```python
plt.plot(lx,ly,c='blue')
plt.scatter(  df_test_negative['Clump Thickness'], df_test_negative['Cell Size'], marker='o', s=200, 
            c='red' )
plt.scatter(  df_test_positive['Clump Thickness'], df_test_positive['Cell Size'], marker='x', s=150, 
            c='black' )
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')

plt.show()
```
